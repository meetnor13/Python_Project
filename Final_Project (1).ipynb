````python
user_name = str(input("Enter your name : ")) # Asks user to enter to the name 
print("Hello",user_name,"! Welcome to Number Guessing Game.") # welcoming the user with his/her name in game
import random # Generating or manuplating the random numbers with the random module
computer_guess = random.randint(1,100) # returns the random numbers from given range 1 to 100
count = 0 # Initializing the counting of iteration of code
#print(computer_guess) # printing the random inetegr which is generated by a computer # just for reference

while True: #initiating loop as user have to guess an integer
    # ask user to enter an integer which is between 1 to 100 or press 0 to quit the game in the middle of the game
    user_guess = int(input("Guess a number between '1' and '100' or Press '0' to quit = ")) 
    count += 1 # counts the number of guesses
    print("Trial",count,'=',user_guess)
   
    if user_guess != 0:    # assuming that user will enter a digit except zero
        if computer_guess != user_guess: # assuming that the entered number by user is not matching the guessed number by a computer
            if computer_guess < user_guess: # comapring the number entered by the user and guessed by a computer
                print("Your guess is too High")
            else:
                print("Your guess is too Low")
            continue # indicating to continue the loop 
        else: # assuming that the entered number by a user is matched with the guessed number by a computer
            print("CONGRATULATIONS! Yoh have guessed a correct number") # telling the user that they have guessed a correct number
            your_score = int(1000 / count) # calculte the score of user
            print(user_name,"score is = ",your_score) # returns the score value
            break  # stopping the loop as user have made a correct guess
    else: # assuming that user enters a zero as he/she has to quit the game
        Quit = str(input("Are you sure you want to Quit? 'Yes' or 'No' = ")) # asking user as are they sure for quitting the game
        if Quit == 'No': # assuming that user does not want to quit the game
            continue # continuing the loop
        elif Quit == 'Yes': # assuming that user want to quit the game
          print("SORRY! You have failed!") # giving feedback to user that user has failed
          print("Bye Bye! Play again!") # telling user to play again
          break # stopping the loop as user has quit the game
        else: # assuming that the user enters a string other than 'Yes' or 'No'
          print("Invalid Input") # indicating the user the entered input or vlaue is invalid
        
````
